```
from bs4 import BeautifulSoup

html ='''
<html>
<body>
<a href='https://www.naver.com' style ='color:blue'>naver</a>
<p>test</p>
</body>
</html>
'''

time_sum=0
soup = BeautifulSoup(html, 'lxml')
```

```
inf=soup.a # 딕셔너리 
print(inf.attrs)
```
