## 네이버 영화

```python
url = 'https://movie.naver.com/movie/running/current.nhn'
page = urlopen(url)
page
```

```python
soup = bs(page, 'html.parser')
print(soup.prettify)
```

```python
s1 = soup.find('div', class_='lst_wrap')
print(s1)
```

```python
s2 = s1.find('ul', class_='lst_detail_t1')
s2
```

```python
s3 = s2.find_all('dl', class_='lst_dsc')
s3
```

```python
len(s3)
```

```python
name = []
for items in s3:
    s4 = items.find('dt', class_='tit')
    s5 = s4.a.text
    name.append(s5)
name    
```

```python
grade =[]
for items in s3:
    s4 = items.find('span', class_='num')
    s5 = s4.text
    grade.append(s5)
grade
```

```python
listDir_name=[]
for items in s3:
    s4 = items.find_all('span', class_='link_txt')
    s5 = s4[1].find_all('a')
    dir_name=''
    for idx in range(len(s5)):
        if idx == 0 :
            dir_name = s5[0].text
            
        else:
            dir_name = dir_name + ',' + s5[1].text
        
    listDir_name.append(dir_name)
print(len(listDir_name))
print(listDir_name)
    
        
        
    
    #print(s5)
```
