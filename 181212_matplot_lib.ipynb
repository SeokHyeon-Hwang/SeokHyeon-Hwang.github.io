# 여러개의 그래프

```python
fig=plt.figure() # 전체 틀을 만든다.
ax1 = fig.add_subplot(2,1,1) # 2행 1열의 첫번째 ax1
ax2 = fig.add_subplot(2,1,2) # 2행 1열의 첫번째 ax1

x=range(0,100)
y=[v*v for v in x]

ax1.plot(x,y)
ax2.bar(x,y)

plt.show()
```

## sin, cos graph

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(0.0, 2*np.pi, 0.1)
sin_y=np.sin(x)
cos_y=np.cos(x)

fig=plt.figure()
ax1 = fig.add_subplot(2,1,1)
ax2 = fig.add_subplot(2,1,2)

ax1.plot(x, sin_y, 'b--')
ax2.plot(x, cos_y, 'r--')

plt.show()
```
