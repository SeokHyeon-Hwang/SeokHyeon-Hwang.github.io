# t-SNE를 이용한 매니폴더(manifold learning)


```python
import matplotlib.pyplot as plt
import matplotlib

from sklearn.datasets import load_digits
```

```python
# 한글
from matplotlib import font_manager, rc
font_loc = "C:/Windows/Fonts/malgunbd.ttf"
font_name = font_manager.FontProperties(fname=font_loc).get_name()
matplotlib.rc('font', family=font_name)
```

```python
digits = load_digits()
digits.data.shape
```

```python
fig, axes = plt.subplots(2, 5, figsize=(10, 5),
                        subplot_kw={'xticks':(), 'yticks':()})
for ax, img in zip(axes.ravel(), digits.images):
    ax.imshow(img)
```

```python
fig, axes = plt.subplots(3, 5, figsize=(10, 5),
                        subplot_kw={'xticks':(), 'yticks':()})
for ax, img in zip(axes.ravel(), digits.images):
    ax.imshow(img)
```

```python
fig, axes = plt.subplots(4, 5, figsize=(10, 5),
                        subplot_kw={'xticks':(), 'yticks':()})
for ax, img in zip(axes.ravel(), digits.images):
    ax.imshow(img)
```

```python
fig, axes = plt.subplots(8, 10, figsize=(20, 10),
                        subplot_kw={'xticks':(), 'yticks':()})
for ax, img in zip(axes.ravel(), digits.images):
    ax.imshow(img)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
pca.fit(digits.data)
# 처음 두 개의 주성분으로 숫자 데이터를 변환한다.
digits_pca = pca.transform(digits.data)
colors = ["#476A2A", "#7851B8", "#BD3430", "#4A2D4E", "#875525",
"#A83683", "#4E655E", "#853541", "#3A3120","#535D8E"]
plt.figure(figsize=(10, 10))
plt.xlim(digits_pca[:, 0].min(), digits_pca[:, 0].max())
plt.ylim(digits_pca[:, 1].min(), digits_pca[:, 1].max())

for i in range(len(digits.data)):
# 숫자 텍스트를 이용해 산점도를 그립니다
    plt.text(digits_pca[i, 0], digits_pca[i, 1], str(digits.target[i]),
    color = colors[digits.target[i]],
    fontdict={'weight': 'bold', 'size': 9})
plt.xlabel("첫 번째 주성분")
plt.ylabel("두 번째 주성분")
```

```python
from sklearn.manifold import TSNE
tsne = TSNE(random_state=42)
d_tsne = tsne.fit_transform(digits.data)
```

```python
print(digits.data.shape)
print(d_tsne.shape)
```

```python
print(d_tsne[:,0].shape)
```

```python
plt.figure(figsize=(12,10))
plt.xlim(d_tsne[:,0].min(), d_tsne[:,0].max()+1) # x축 범위 설정
plt.ylim(d_tsne[:,1].min(), d_tsne[:,1].max()+1) # y축 범위 설정

for i in range(len(digits.data)):
    plt.text(d_tsne[i,0],
            d_tsne[i,1],
            str(digits.target[i]),
            color = colors[digits.target[i]],
            
            )
```

```python
from sklearn import datasets
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
iris = datasets.load_iris()
iris.data.shape
pca.fit(iris.data)
# 처음 두 개의 주성분으로 숫자 데이터를 변환한다.
iris_pca = pca.transform(iris.data)
colors = ["#476A2A", "#7851B8", "#BD3430", "#4A2D4E", "#875525",
"#A83683", "#4E655E", "#853541", "#3A3120","#535D8E"]
plt.figure(figsize=(10, 10))
plt.xlim(iris_pca[:, 0].min(), iris_pca[:, 0].max())
plt.ylim(iris_pca[:, 1].min(), iris_pca[:, 1].max())

for i in range(len(iris.data)):
# 숫자 텍스트를 이용해 산점도를 그립니다
    plt.text(iris_pca[i, 0], iris_pca[i, 1], str(iris.target[i]),
    color = colors[iris.target[i]],
    fontdict={'weight': 'bold', 'size': 9})
plt.xlabel("첫 번째 주성분")
plt.ylabel("두 번째 주성분")
```

```python
from sklearn.manifold import TSNE
tsne = TSNE(random_state=42)
d_tsne = tsne.fit_transform(iris.data)
```
