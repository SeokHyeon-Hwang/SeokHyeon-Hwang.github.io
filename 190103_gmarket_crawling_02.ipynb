```python
from bs4 import BeautifulSoup
import requests as req
import html5lib
```

```python
url = 'http://search.gmarket.co.kr/search.aspx?selecturl=total&sheaderkey=&gdlc=&SearchClassFormWord=goodsSearch&keywordOrg=%BF%A9%BC%BA%C0%C7%B7%F9&keywordCVT=%BF%A9%BC%BA%C0%C7%B7%F9%2C%BA%F2%BB%E7%C0%CC%C1%EE%BF%A9%BC%BA%C0%C7%B7%F9%2C%C1%DF%B3%E2%BF%A9%BC%BA%C0%C7%B7%F9&keywordCVTi=1&keyword=%BF%A9%BC%BA%C0%C7%B7%F9'
res = req.get(url)
html = res.content
soup = BeautifulSoup(html, 'html5lib')
print(soup.prettify())
```

```python
soup_item = soup.find('div', {'id' :'divListAjaxContainer'})
print(soup_item.prettify())
```

```python
soup_it2 = soup_item.find('ul', {'id':'searchListItems'})
print(soup_it2)
```

```python
soup_it3 = soup_it2.find_all('li')
print(soup_it3)
```

```python
titlel = []
pricel = []
```

```python
soup_it3 = soup_it2.find_all('li')
for i in soup_it3:
    title = i.find('span', {'class':'title'}).text
    price = i.find('span', {'class':'price'}).find('strong').text
    titlel.append(title)
    pricel.append(price)
    print(title)
    print(price)
```

```python
print(len(titlel))
print(len(pricel))
```

```python
print(titlel, pricel)
```
