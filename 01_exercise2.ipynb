```
### lxml, html5의 시간 비교
import time
```

```
startTime=time.time()

print('시간측정')

endTime =time.time() = startTime
print(endTime)
```

```
html = """
<html>
<head>
</head>
<body>
<p>test</p>
</body>
</html>
"""

start_time1=time.time() # 시작 시간을 저장
BeautifulSoup(html, 'lxml')
endTime1 =time.time() - start_time1


start_time2=time.time() # 시작 시간을 저장
BeautifulSoup(html, 'html5lib')
endTime2 =time.time() - start_time2


print(endTime1, endTime2)
```
