```python
url = 'https://datalab.naver.com/keyword/realtimeList.naver'
page = urlopen(url)
soup = bs(page, 'html.parser')
print(soup.prettify)
```

```python
from selenium import webdriver
driver = webdriver.Chrome('chromedriver') # 웹 브라우저 띄우기
driver.get('https://datalab.naver.com/keyword/realtimeList.naver')
```

```python
soup = bs(driver.page_source, 'html.parser')
s1 = soup.find_all('span', class_='title')
s1
```

```python
s1 = soup.find_all('ul', class_='rank_list v2')
s1
len(s1)
```

```python
s2 = s1[1].find_all('span', class_='title')
s2
```

```python
keyw10=[]
keyw20=[]
keyw30=[]
keyw40=[]
keyw50=[]
keywall=[]

cnt=0
for items in s1:
    keyw_l = items.find_all('span', class_='title')
    for items2 in keyw_l:
        if cnt==0:
            keywall.append(items2.text)
        elif cnt==1:
            keyw10.append(items2.text)
        elif cnt==2:
            keyw20.append(items2.text)
        elif cnt==3:
            keyw30.append(items2.text)
        elif cnt==4:
            keyw40.append(items2.text)
        elif cnt==5:
            keyw50.append(items2.text)
    cnt+=1
    
print(keyw10)
        
        
```

```python
dict_val = {'전체':keywall, '10대':keyw10, '20대':keyw20, '30대':keyw30, '40대':keyw40, '50대':keyw50}
```

```python
import pandas as pd
dat = pd.DataFrame(dict_val)
dat
```

```python
dat.to_excel('naver_trend.xlsx', sheet_name='Keyword', index=False)
```
