```
from sklearn.datasets import make_moons
from sklearn.cluster import KMeans

X, y = make_moons(n_samples=200, noise=0.05, random_state=0)
X.shape
```

```
y.shape
```

```
import matplotlib.pyplot as plt
```

```
kmeans = KMeans(n_clusters=3)
kmeans.fit(X)
y_pred = kmeans.predict(X)
y_pred
```

```
col1 = X[ : , 0 ]  # 1
col2 = X[ : , 1 ]
plt.scatter(col1, col2, c=y_pred, cmap=mglearn.cm3, edgecolors='g')
```

```
##
kmeans = KMeans(n_clusters=2)
kmeans.fit(X)
y_pred = kmeans.predict(X)
y_pred
```

```
col1 = X[ : , 0 ]  # 1
col2 = X[ : , 1 ]
plt.scatter(col1, col2, c=y_pred, cmap=mglearn.cm2, edgecolors='g')
```

```
print(kmeans.cluster_centers_)
```

```
kmeans = KMeans(n_clusters=2)
kmeans.fit(X)
y_pred = kmeans.predict(X)
y_pred


col1 = X[ : , 0 ]  # 1
col2 = X[ : , 1 ]

## 
centerX = kmeans.cluster_centers_[  : , 0  ]
centerY = kmeans.cluster_centers_[  : , 1  ]

plt.scatter(col1, col2, c=y_pred, cmap=mglearn.cm2, edgecolors='g')
plt.scatter(centerX,  centerY, marker="^")
```

```
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
```

```
X_varied, y_varied = make_blobs(n_samples=200,
                                cluster_std=[1.0, 2.5, 0.5],
                                random_state=170)
```

```
X_varied
```

```
y_pred = KMeans(n_clusters=3, random_state=0).fit_predict(X_varied)
y_pred
```

```
mglearn.discrete_scatter(X_varied[: , 0],
                         X_varied[: , 1], 
                         y_pred)
```

```
###
from sklearn.decomposition import PCA #   

import seaborn as sns
%matplotlib inline
```

```
iris = sns.load_dataset('iris')
iris.head()
print(type(iris))
X_iris = iris.drop('species', axis=1)
y_iris = iris['species']
X_iris
```

```
model = PCA(n_components=2) #   .   
model.fit(X_iris) # y  .  
X_2D = model.transform(X_iris) #  2 
```

```
print(X_iris.head())
print(X_2D)
```

```
X_2D[: , 0] #   
```

```
iris['PCA1'] = X_2D[: ,0] # feature 
iris['PCA2'] = X_2D[: ,1] # feature 
sns.lmplot("PCA1", "PCA2", hue="species", data=iris, fit_reg=False);
```
