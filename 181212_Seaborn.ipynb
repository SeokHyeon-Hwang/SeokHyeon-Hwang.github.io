```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# plt.show()를 생략해 주는 주피터 명령
%matplotlib inline 
```

```python
import seaborn as sns
```

```python
sns.set_style('whitegrid')
```

```python
x=np.linspace(0,14,100) # 0~14 100조각
y1 = np.sin(x)
y2 = 2*np.sin(x)
y3 = 4*np.sin(x)
```

```python
plt.figure(figsize=(10,6) ) # 그래프 크기
plt.plot(x, y1)
plt.plot(x, y2)
plt.plot(x, y3)

```

```python
# 1행 3렬 만들기
x=np.linspace(0,14,100)
y1 = np.sin(x)
y2 = 2*np.sin(x)
y3 = 4*np.sin(x)

fig=plt.figure(figsize=(10,6))
ax1 = fig.add_subplot(1,3,1)
ax2 = fig.add_subplot(1,3,2)
ax3 = fig.add_subplot(1,3,3)

ax1.plot(x, y1, 'r--')
ax2.plot(x, y2, 'gs-')
ax3.plot(x, y3, 'bo-')
```

## 항공데이터

```python
fg = sns.load_dataset('flights')
print(fg.shape, fg.columns)
fg.head(5)
```

```python
fgp = fg.pivot('month', 'year', 'passengers')
fgp
plt.figure(figsize=(10,8))
sns.heatmap(fgp, annot=True, fmt='d') # 글자를 쓰고 정수값 형태로
plt.show()
```
