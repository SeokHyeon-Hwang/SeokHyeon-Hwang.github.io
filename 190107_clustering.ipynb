```
from sklearn.datasets import make_moons
from sklearn.cluster import KMeans

X, y = make_moons(n_samples=200, noise=0.05, random_state=0)
X.shape
```

```
y.shape
```

```
import matplotlib.pyplot as plt
```

```
kmeans = KMeans(n_clusters=3)
kmeans.fit(X)
y_pred = kmeans.predict(X)
y_pred
```

```
col1 = X[ : , 0 ]  # 1열선택
col2 = X[ : , 1 ]
plt.scatter(col1, col2, c=y_pred, cmap=mglearn.cm3, edgecolors='g')
```

```
##
kmeans = KMeans(n_clusters=2)
kmeans.fit(X)
y_pred = kmeans.predict(X)
y_pred
```

```
col1 = X[ : , 0 ]  # 1열선택
col2 = X[ : , 1 ]
plt.scatter(col1, col2, c=y_pred, cmap=mglearn.cm2, edgecolors='g')
```

```
print(kmeans.cluster_centers_)
```

```
kmeans = KMeans(n_clusters=2)
kmeans.fit(X)
y_pred = kmeans.predict(X)
y_pred


col1 = X[ : , 0 ]  # 1열선택
col2 = X[ : , 1 ]

## 중앙점
centerX = kmeans.cluster_centers_[  : , 0  ]
centerY = kmeans.cluster_centers_[  : , 1  ]

plt.scatter(col1, col2, c=y_pred, cmap=mglearn.cm2, edgecolors='g')
plt.scatter(centerX,  centerY, marker="^")
```

```
###
from sklearn.decomposition import PCA # 모델 클래스 선택

import seaborn as sns
%matplotlib inline
```

```
iris = sns.load_dataset('iris')
iris.head()
print(type(iris))
X_iris = iris.drop('species', axis=1)
y_iris = iris['species']
X_iris
```

```
model = PCA(n_components=2) # 유지할 구성 요소수. 없으면 전체 선택
model.fit(X_iris) # y는 지정 안함. 데이터 학습
X_2D = model.transform(X_iris) # 데이터를 2차원으로 변환
```

```
print(X_iris.head())
print(X_2D)
```

```
X_2D[: , 0] # 첫번째 열 데이터
```

```
iris['PCA1'] = X_2D[: ,0] # feature 생성
iris['PCA2'] = X_2D[: ,1] # feature 생성
sns.lmplot("PCA1", "PCA2", hue="species", data=iris, fit_reg=False);
```
