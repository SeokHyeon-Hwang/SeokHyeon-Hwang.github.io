### Cell을 Run하는 방법
＊Shift+Enter:현재 cell을 Run하고 다음 cell로 넘어간다. 만약 다음 cell이 없다면 다음 cell을 만든다.

＊ctrl+Enter:현재 cell을 실행하고 다음 cell로 커서가 넘어가지 않는다.

＊Alt+Enter:현재 cell을 Run하고, 다음 cell로 넘어간다. Cell 사이에 실행했다면, 중간에 새로운 cell을 삽입한다.

### Cell을 Split 하는 방법
＊ctrl+shift+'-' : 나누고자 하는 위치에서.
