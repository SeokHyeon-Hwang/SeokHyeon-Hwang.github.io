## 01. Tensorflow에서의 변수 표현

```
import tensorflow as tf
```

```
# 상수 정의할 때, 
a1 = tf.constant(10, name="a")
b1 = tf.constant(20, name="b")
c1 = tf.constant(30, name="c")
```

```
# 변수 정의하기 
v = tf.Variable(0, name='v')  # 변수 v선언하고 초기값 0
```

```
# 데이터 플로우 그래프 정의, 그래프를 실행
calc_op = a1 + b1 + c1
assign_op = tf.assign(v, calc_op) # v에 calc_op 대입
```

```
# 세션을 실행
sess = tf.Session()
sess.run(assign_op)
```

```
sess.run(v)
```

## 02. 플레이스 홀더(Placeholder)

* 템플릿처럼 값을 넣을 공간을 만들어둔다.
* 플레이스 홀더를 이용하면 데이터 플로우 그래프를 구축시에 값을 넣지 않고, 
* 세션을 실행시에 값을 넣을 수 있다.

```
# 플레이스 홀더 정의하기
a = tf.placeholder(tf.int32, [3])  # 정수 자료형 3개를 가진 배열

# 배열을 모든 값을 2배한다.
b = tf.constant(2)
x2_op = a * b
```

```
# 세션을 시작
sess = tf.Session() 
# 플레이스 홀더에 값을 넣고 실행.
r1 = sess.run(x2_op, feed_dict={ a:[1,2,3]    })
r2 = sess.run(x2_op, feed_dict={ a:[10,20,30] })
print(r1,r2)
```
